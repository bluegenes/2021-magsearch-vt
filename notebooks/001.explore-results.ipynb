```python
ms31 = pd.read_csv("../output.magsearch/results/pl-genomes.k31.csv.gz",
                   sep=",",
                   quotechar="'",
                   header=0,
                   names=["MAG", "metagenome", "containment"])

# Fix names so it's easier to query
ms31['MAG'] = ms31['MAG'].str.replace(r"'(?P<id>.*)'", lambda m: m.group("id"), regex=True)
ms31['accession'] = ms31['MAG'].str.split(" ", 1, expand=True)[0]
ms31['sci_name'] = ms31['MAG'].str.split(" ", 1, expand=True)[1]

ms31['metagenome'] = ms31['metagenome'].str.replace(r".*/(?P<id>.*).sig.*", lambda m: m.group("id"), regex=True)

ms31
```
